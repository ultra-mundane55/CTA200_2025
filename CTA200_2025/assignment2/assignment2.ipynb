```python
import numpy as np
```

```python
"""
This function calculates x^3-x^2-1
Parameters: x 
Returns: the quantity x^3-x^2-1
"""
def f(x):
    return (x**3)-(x**2)-1

"""
This function calculates 3x^-2x, the derivative 
of x^3-x^2-1
Parameters: x 
Returns: the quantity 3x^-2x
"""
def df(x):
    return 3*(x**2)-2*x
```

### Part 2

$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$, stop when $|f(x_n)| < \epsilon$

```python
"""
This function calculates the Newtonian Iteration of 
the function f with derivative df.
Parameters: 
    f:  function
        input function 
    df: function
        derivative of f
    x0: float
        idk
    epsilon: float
             idk
    max_iter: float
              idk
Returns: 

    xn: float
        the root x_n such that f(x_n)=0, or 
    None:
        if the function fails

"""

def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    xn=x0
    for i in range(max_iter):
        if np.abs(f(xn))<epsilon:
            print(f"Found root in {i} iterations")
            # print(f"root is {xn}")
            return xn
        xn = xn-(f(xn)/df(xn))
    print ("Iteration failed.")

            
```

### Part 3

```python
# 2 different values for x0
newton(f, df, 8)
newton(f, df, 2)

# reduce epsilon
newton(f, df, 2, epsilon=1e-8)

# took one more iteration to converge than 1e-6
```
